If insured members cliams incline at the same progression, lets forecast.

```python
insurance_data
```

```python
insurance_data.shape
```

```python
import statsmodels
from statsmodels.tsa.ar_model import AutoReg
aic_score = []
for index in range(1):
    model = AutoReg(insurance_data["claim"], lags= index )
    model_fitted = model.fit()
    aic_score.append(model_fitted)
```

```python
model = AutoReg(insurance_data["claim"], lags= 1)
model_fitted = model.fit()
print(model_fitted.summary())
```

```python
fcast = model_fitted.predict(start = len(insurance_data),end = len(insurance_data) + 100)
```
