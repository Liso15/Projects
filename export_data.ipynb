```python
# importation and libraries
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
export_data = pd.read_csv(pwd + "/exportacao_full.csv", index_col=[0], parse_dates=True)
export_data
```

```python
# all the columns in the dataset

list(export_data.columns)
```

```python
# check all the distint value within the columns, we want to get a picture of what kind of dataset we are dealing with:
export_data["Country"].unique()
```

```python
# export destinetion cities
list(export_data["City"].unique())
```

```python
# Economy exported 
list(export_data["Economic Block"].unique())
```

```python
# distint consumbles exported

export_data["SH2 Description"].unique()
```

```python
# basic descriptive statistics in the $-column
export_data["US$ FOB"].describe()
```

```python
# filter European exports

export_data[export_data["Economic Block"] == "Europe"]
# already we understand that most exports recorded in the dataset where in European countries or euro-area
```

```python
# European union
export_data[export_data["Economic Block"] == "European Union (EU)"]
```

```python
# Droping columns SH4 Code and SH4 Description
export_data.drop( columns="SH4 Code", inplace=True)
export_data
```

```python
# Droping Net weight column and SH2 Code

export_data.drop(columns="SH2 Code", inplace=True)
export_data
```

```python
export_data.drop(columns="SH4 Description", inplace=True)
export_data
```

```python
export_data.drop(columns="Net Weight", inplace=True)
export_data
```

```python
#Duplicated values

export_data.drop_duplicates(["SH2 Description"]).reset_index()
export_data.duplicated(["SH2 Description"]).reset_index()
```

```python
export_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# grouby Country by $ they retured export
export_data.groupby("Country")["US$ FOB"].sum().reset_index()
#Sort by US$ FOB
export_data.groupby("Country")["US$ FOB"].sum().reset_index().sort_values(by="US$ FOB", ascending=False)
```

```python
export_data.groupby("Country")["US$ FOB"].sum().reset_index().sort_values(by="US$ FOB", ascending=False).plot.hist(bins=50 )
```

```python
# which city was exported the most?: grouby City by SH2 Description
export_data.groupby("City")["SH2 Description"].count().reset_index()
#Sort by SH2 Description
export_data.groupby("City")["SH2 Description"].count().reset_index().sort_values(by="SH2 Description", ascending=False)
```

```python
#In which Economic Block in $ was returned the most?: groupby Economic Block and $
export_data.groupby("Economic Block")["US$ FOB"].sum().reset_index()
#sort US$ FOB column
export_data.groupby("Economic Block")["US$ FOB"].sum().reset_index().sort_values(by="US$ FOB", ascending=False)
```

```python
sns.barplot(export_data.groupby("Economic Block")["US$ FOB"].sum().reset_index().sort_values(by="US$ FOB", ascending=False), x= 'US$ FOB', y='Economic Block', orient='h')
```

```python
# groupby Year relation to US$ FOB

export_data.groupby("Year")["US$ FOB"].sum().reset_index().sort_values(by= "Year", ascending=True)
```

```python

sns.lineplot(export_data.groupby("Year")["US$ FOB"].sum().reset_index().sort_values(by= "Year", ascending=True), x='Year', y='US$ FOB', color='k', dashes=False)
```

```python

from pandas import pivot_table
# pivot_table showing revenue per country in years:

pivot_table = export_data.pivot_table(index="Country", columns="Year", values="US$ FOB", aggfunc='sum', fill_value=0).reset_index()
pivot_table
```

```python
# exports with H2 Description per Country

pivot_table = export_data.pivot_table(index="Country", columns="Year", values="SH2 Description", aggfunc='count', fill_value=0).reset_index()
pivot_table
```
